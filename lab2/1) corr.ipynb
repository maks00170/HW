```python
comment = comment[['ID поста', 'Эмоциональный окрас']]
```

```python
report = report[['ID поста', 'Эмоциональный окрас']]
```

```python
report
```

```python
report.rename( columns = {'Эмоциональный окрас': 'Эмоциональный окрас поста'}, inplace = True);
```

```python
df = report.merge(comment, on = 'ID поста')
df
```

```python
piv_df = pd.pivot_table(df, values='Эмоциональный окрас поста', index='ID поста', columns='Эмоциональный окрас', aggfunc='count')
```

```python
piv_df = piv_df.fillna(0)
```

```python
piv_df
```

```python
piv_df = piv_df.merge(report, on = 'ID поста')
```

```python
piv_df
```

```python
piv_df['corr_n'] = -1
piv_df['corr_n_k'] = -1
piv_df['corr_n_c'] = -1
```

```python
col = ['Вежливость', 'Негатив', 'Нейтральность', 'Неопределенность', 'Позитив', 'Юмор']
k = 1
corr_n = []
corr_n_k = []
corr_n_c = []
for ID in piv_df['ID поста']:
    mask = piv_df['ID поста'] == ID
    n = int(np.sum(piv_df[mask][col], axis = 1))
    sint = piv_df[mask]['Эмоциональный окрас поста'].values[0]
    cor_n = piv_df[mask][sint].values[0]/n
    cor_n_k = piv_df[mask][sint].values[0]/(n+k)
    cor_n_c = 2*piv_df[mask][sint].values[0]/(n+piv_df[mask][sint].values[0])
#     if n <= k:
#         cor_n_1 = -1
        
    corr_n.append(cor_n)
    corr_n_k.append(cor_n_k)
    corr_n_c.append(cor_n_c)
```

```python
piv_df['corr_n'] = corr_n
```

```python
piv_df['corr_n_k'] = corr_n_k
```

```python
piv_df['corr_n_c'] = corr_n_c
```

```python
piv_df[piv_df['corr_n'] !=0].sort_values(by = ['corr_n_c'], ascending = False)
```
